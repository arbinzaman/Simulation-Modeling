```python
def chi_square(o,e):
    alpha =0.05
    df = 3
    rejection_region = 7.81
    x2 = 0
    for item1, item2 in zip(o,e):
        x2 += ((item1-item2)**2)/item2
        print("Value of x2 is ",x2)
        if x2 > rejection_region:
            print("Null hypothesis is rejected")
        else:
            print("Null hypothesis is accepted")
```
